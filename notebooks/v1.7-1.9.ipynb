## Version 1.7

This notebook marks a new chapter in the `utility-sim-tools` project.

While versions **1.0 â€“ 1.6** were focused entirely on the **1D heat equation pipeline**, this version introduces two important new capabilities:

> This prepares the project for public use, while keeping the simulation core untouched.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
sys.path.append(os.path.abspath("../src"))

from initial_conditions.gaussian_2d import gaussian_bump_2d

L = 10.0
N = 129
x = np.linspace(-L/2, L/2, N)
X, Y = np.meshgrid(x, x)

Z = gaussian_bump_2d(X, Y, center=(0.0, 0.0), width=1.0, amplitude=1.0)

plt.figure(figsize=(6, 5))
plt.contourf(X, Y, Z, levels=50, cmap="viridis")
plt.colorbar(label="Amplitude")
plt.title("2D Gaussian Bump")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
plt.savefig("figures/gaussian_2d_preview.png", dpi=300)
plt.show()
```

###  Streamlit GUI Mockup

We created a simple graphical interface (`mock_gui.py`) using [Streamlit](https://streamlit.io/). It allows the user to:
